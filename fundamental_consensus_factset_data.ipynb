# How to look up COMPANY ID and tickers for FACTSET Data
This sample show how to achieve the following operation with a simple interface:
- Look up the unique FACTSET instrument ID (FSYM_ID) from a symbols, package and freq

## `DAL`  Data Accessor Library 
> DAL access data from FACTSET database / API for Equity Data  ## Import ASR Data Accessor Library (DAL)

```python
from aiq_strategy_robot.data.data_accessor import DAL
handler = DAL()
data_source='FACTSET'
```

### Pattern 1: look up FACTSET from symbols, package and freq.

####  **load**
> Load the data to memory  
> Set data_type = 'symbol_lookup' for looking up instrument IDs.

```python
# `data_type` must be 'fundamental_estimates'
handler = handler.load(
    data_source, # type: ignore
    data_type="fundamental_estimates",
    symbols="R2HXLJ-R",
    # fields=["SALES", "EPS"],
    package="BASIC",
    freq=3,
    # start_datetime="2021-01-01",
    # end_datetime= "2023-06-30"
)
```

#### **Retrieve**
> `Pandas.DataFrame` the loaded data via `retrieve`

```python
df_symbol = handler.retrieve(transformed=False)
df_symbol.head(20)
```

#### Too many results were retrieved.
> The search result can be made shorter by specifying fields, start_datetime and/or end_datetime.

```python
# This time currency_code and exchange_code are specified.
# pick_only_primary can be also used to focus on only the main exchange.
# `data_type` must be 'quants_factor_data'
handler = handler.load(
    data_source, # type: ignore
    data_type="fundamental_estimates",
    symbols="R2HXLJ-R",
    fields=["SALES", "EPS"],
    package="BASIC",
    freq=3,
    start_datetime="2021-01-01",
    end_datetime= "2023-06-30"
)
df_symbol = handler.retrieve(transformed=False)
df_symbol.head(20)
```

#### Caution: You must input symbols package and freq in searching.

かくなら、symbols,package,freq,daytypeは省略できない

```python
# An error is thrown if you do not input exactly one of either symbols, package, or freq when searching.
handler = handler.load(
    data_source, # type: ignore
    data_type="fundamental_estimates",
    # symbols="R2HXLJ-R",
    fields=["SALES", "EPS"],
    package="BASIC",
    freq="CONH",
    start_datetime="2021-01-01",
    end_datetime= "2023-06-30"
)
```
