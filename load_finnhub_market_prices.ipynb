## Appendix 

> Handler.load()

Can also be obtained directly from the handler without using the wrapper.<br>
In this case, `data_source` and `data_type` must be specified.

```python
sdh.load(
    'finnhub', 
    data_type='equity_prices',
    symbol='8035.T', 
    freq='60', 
    start_datetime='2023-12-29', 
    end_datetime='2024-02-01'
).retrieve()
```

```python
sdh.load(
    'finnhub', 
    data_type='dividends',
    symbol='8035.T',
    start_datetime='2015-09-30', 
    end_datetime='2023-12-31'
).retrieve()
```

```python
sdh.load(
    'finnhub', 
    data_type='stock_splits',
    symbol='8035.T',
    start_datetime='2015-09-30', 
    end_datetime='2023-12-31'
).retrieve()
```

```python
sdh.load(
    'finnhub', 
    data_type='equity_data',
    symbols=['AMZN', '8035.T'],
    freq='D',
    start_datetime='2015-09-30', 
    end_datetime='2024-01-01'
).retrieve()
```
