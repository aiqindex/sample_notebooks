# Appendix 

> Handler.load()

Can also be obtained directly from the handler without using the wrapper.<br>
In this case, `data_source` and `data_type` must be specified.

```python
sdh.load(
    'finnhub', 
    data_type='exchange_code',
    country_name='japan'
).retrieve()
```

```python
sdh.load(
    'finnhub', 
    data_type='symbol_lookup',
    symbols=['4661.T', '8035.T'],
    exchange_code='T'
).retrieve()
```

```python
load_finnhub_symbol_lookup(
    sdh,
    company_names=['drive shack'],
    exchange_code='US'
).retrieve()
```
