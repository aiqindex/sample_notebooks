# How to look up financial Data for finnhub Data

```python
import os
import pandas as pd
from aiq_strategy_robot.data.data_accessor import DAL
```

```python
## load wrapper
from aiq_strategy_robot.data.FINNHUB import load_finnhub_fundamental
```

```python
sdh = DAL()
```

## Get company financials from finnhub symbols

```python
load_finnhub_fundamental(
    sdh,
    symbols='7203.T', 
    st_type='ic', 
    freq='quarterly', 
    start_datetime='2022-01-01', 
    with_calendar=True
).retrieve()
```

```python
load_finnhub_fundamental(
    sdh,
    symbols='8035.T', 
    st_type='bs', 
    freq='quarterly'
).retrieve()
```

```python
load_finnhub_fundamental(
    sdh,
    symbols='8035.T', 
    st_type='cf', 
    freq='quarterly',
    with_calendar=False
).retrieve()
```

```python
load_finnhub_fundamental(
    sdh,
    symbols='AMZN', 
    st_type='ic', 
    freq='quarterly',
    start_datetime='2022-03-31',
    with_calendar=True
).retrieve()
```

```python
load_finnhub_fundamental(
    sdh,
    symbols=['7203.T', '8035.T'], 
    st_type='cf', 
    freq='quarterly',
    with_calendar=True
).retrieve()
```

## Appendix 

> Handler.load()

Can also be obtained directly from the handler without using the wrapper.<br>
In this case, `data_source` and `data_type` must be specified.

```python
sdh.load(
    'finnhub', 
    data_type='fundamental',
    symbols='7203.T', 
    st_type='ic', 
    freq='quarterly', 
    start_datetime='2022-01-01', 
    with_calendar=True
).retrieve()
```
